```python
#Getting all Job Links
links = []
No_Pages = int(input("How many pages you want to scrape: "))
try:
    for page in range(2,No_Pages+3):
        time.sleep(5)
        jobs_window = driver.find_element(By.CLASS_NAME,'jobs-search-results-list')
        jobs_list = jobs_window.find_elements(By.CSS_SELECTOR,".jobs-search-results__list-item")
        for job in jobs_list:
            all_links = job.find_elements(By.TAG_NAME,'a')
            for a in all_links:
                if str(a.get_attribute('href')).startswith("https://www.linkedin.com/jobs/view") and a.get_attribute('href') not in links: 
                    links.append(a.get_attribute('href'))
                else:
                    pass
                driver.execute_script("arguments[0].scrollIntoView();", job)
        page=page-1
        driver.find_element(By.XPATH,f"//button[@aria-label='Page {page}']").click()
        time.sleep(5)
except:
    pass
print('Found ' + str(len(links)) + ' links for job offers')
```

```python
# Create empty lists to store information
Position = []
Company_names = []
Company_locations = []
Working_model = []
Post_dates = []
Role_Type = [] 
Job_desc = []

i=0
j=1
for i in range(len(links)):
    try:
        driver.get(links[i])
        i=i+1
        WebDriverWait(driver,100).until(EC.presence_of_element_located((By.XPATH,'/html/body/div[5]/div[3]/div/div[1]/div[1]/div/div[1]/div/div/div[1]/div[3]/div/button')))
        # Click See more Button
        driver.find_element(By.XPATH,'//*[@id="ember32"]/span').click()
        time.sleep(2)
        
    except:
        pass

    contents = driver.find_elements(By.CLASS_NAME,'p5')
    for content in contents:
        try:
            Position.append(content.find_element(By.TAG_NAME,"h1").text)
            Company_names.append(content.find_element(By.CLASS_NAME,"jobs-unified-top-card__company-name").text)
            Company_locations.append(content.find_element(By.CLASS_NAME,"jobs-unified-top-card__bullet").text)
            Working_model.append(content.find_element(By.CLASS_NAME,"jobs-unified-top-card__workplace-type").text)
            Post_dates.append(content.find_element(By.CLASS_NAME,"jobs-unified-top-card__posted-date").text)
            Role_Type.append(content.find_element(By.CLASS_NAME,"jobs-unified-top-card__job-insight").text)
            print(f'Scraping the Job Offer {j} DONE.')
            j+= 1
            
        except:
            pass
        time.sleep(2)
    
    job_description = driver.find_elements(By.CLASS_NAME,'jobs-description__content')
    for description in job_description:
        job_text = description.find_element(By.CLASS_NAME,"jobs-box__html-content").text
        Job_desc.append(job_text)
        print(f'Scraping the Job Offer {j}')
        time.sleep(3)
```

```python
# Creating the dataframe 
df = pd.DataFrame(list(zip(Position,Company_names,
                    Company_locations,Working_model,
                    Post_dates,Role_Type)),
                    columns =['Position', 'Company_name',
                           'Company_location','Working_model',
                           'Post_date','Role_Type'])

```
